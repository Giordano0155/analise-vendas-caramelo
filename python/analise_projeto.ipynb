```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:,.2f}'.format
```

```python
dados_caramelo = 'CARAMELO_PROJETO.csv'
df_caramelo = pd.read_csv(dados_caramelo, sep= ';')
df_caramelo.head(3)
```

```python
df_caramelo.info()
```

```python
df_caramelo.describe()
```

```python
df_caramelo.columns
```

```python
df_caramelo['Valor de Venda']
```

```python
df_caramelo.groupby('Loja')['Valor de Venda'].sum().reset_index()
df_caramelo = df_caramelo.sort_values(by = 'Valor de Venda', ascending = False)
df_caramelo
```

```python
df_caramelo['Valor de Venda'] = (
    df_caramelo['Valor de Venda']
    .str.replace('.', '', regex=False)   
    .str.replace(',', '.', regex=False)  
    .astype(float)
)
```

```python
df_caramelo['Valor de Venda'].dtype
```

```python
q1 = df_caramelo['Valor de Venda'].quantile(0.25)
q3 = df_caramelo['Valor de Venda'].quantile(0.75)
iqr = q3 - q1
outliers = df_caramelo[
    (df_caramelo['Valor de Venda'] < q1 - 1.5 * iqr)|
    (df_caramelo['Valor de Venda'] > q3 + 1.5 * iqr)
]
```

```python
outliers.head()
```

```python
outliers.shape
```

```python
outliers_por_loja = outliers.groupby('Loja')['Valor de Venda'].count().reset_index()
outliers_por_loja = outliers_por_loja.sort_values(by = 'Valor de Venda',ascending = False)
outliers_por_loja.head()
```

```python
out_liers_por_categoria = outliers.groupby('Categoria Produto')['Valor de Venda'].count().reset_index()
out_liers_por_categoria = out_liers_por_categoria.sort_values(by = 'Valor de Venda', ascending = False)
out_liers_por_categoria
```

```python
vendas_loja_categoria = df_caramelo.groupby(['Loja', 'Categoria Produto'])['Valor de Venda'].sum().reset_index()
vendas_loja_categoria = vendas_loja_categoria.sort_values(['Categoria Produto', 'Valor de Venda'], ascending=[True, False])
vendas_loja_categoria.head(10)
```

```python
top_lojas_categoria = vendas_loja_categoria.groupby('Categoria Produto').first().reset_index()
top_lojas_categoria
```

```python
vendas_produtos = df_caramelo.groupby(['Loja', 'Categoria Produto', 'Produto'])['Valor de Venda'].sum().reset_index()
vendas_produtos = vendas_produtos.sort_values(['Categoria Produto', 'Valor de Venda'], ascending=[True, False])

vendas_produtos.head(10)
```

```python
total_vendas = df_caramelo['Valor de Venda'].sum()
total_outliers = outliers['Valor de Venda'].sum()
percentual_outliers = (total_outliers / total_vendas) * 100
print(f"{percentual_outliers:.2f}% das vendas são outliers")
```

```python
df_caramelo['Data da Compra'] = pd.to_datetime(df_caramelo['Data da Compra'])
```

```python
df_caramelo['Ano'] = df_caramelo['Data da Compra'].dt.year
df_caramelo['Mes'] = df_caramelo['Data da Compra'].dt.month
df_caramelo['Ano_Mes'] = df_caramelo['Data da Compra'].dt.to_period('M')
```

```python
vendas_tempo = df_caramelo.groupby('Ano_Mes')['Valor de Venda'].sum().reset_index()

vendas_tempo['Ano_Mes'] = vendas_tempo['Ano_Mes'].astype(str)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(vendas_tempo['Ano_Mes'], vendas_tempo['Valor de Venda'])
plt.xticks(rotation=45)
plt.title('Evolução das Vendas ao Longo do Tempo')
plt.xlabel('Período')
plt.ylabel('Valor de Venda')
plt.show()
```

```python
produto_tempo = df_caramelo.groupby(['Ano_Mes', 'Produto'])['Valor de Venda'].sum().reset_index()
```

```python
top_produto = df_caramelo.groupby('Produto')['Valor de Venda'].sum().idxmax()

produto_especifico = produto_tempo[produto_tempo['Produto'] == top_produto]
```

```python
plt.figure()
plt.plot(produto_especifico['Ano_Mes'].astype(str), produto_especifico['Valor de Venda'])
plt.xticks(rotation=45)
plt.title(f'Evolução de vendas - {top_produto}')
plt.show()
```
